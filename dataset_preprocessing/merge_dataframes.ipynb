```python
df_normal = pd.read_csv("kairos_normal.csv")
df_normal1 = pd.read_csv("kairos_normal_1.csv")

df_window = pd.read_csv("kairos_window.csv")
df_window1 = pd.read_csv("kairos_window_1.csv")
```

```python
df_normal_final = pd.concat([df_normal, df_normal1],ignore_index=True)
df_normal_final= df_normal_final.iloc[: , 1:]
df_normal_final.to_csv("kairos_normal_final.csv")
df_normal_final
```

```python
df_window_final = pd.concat([df_window, df_window1],ignore_index=True)
df_window_final= df_window_final.iloc[: , 1:]
df_window_final.to_csv("kairos_window_final.csv")
df_window_final
```
